# Spread of Influence Analysis View

This file shows the applications of all the diffusion patterns of influence chosen for the test. For both models, all threshold calculation strategies are taken into consideration, tested and analyzed if the results obtained correspond to the hypotheses initially made.

```python
import warnings
warnings.filterwarnings("ignore")

import sys 

sys.path.append("..")
sys.path.append("../../graph")
sys.path.append("../../metrics")

import graph as gr
import metrics as me
import spread_of_influence as sp

import pandas as pd
import numpy as np

ds1_path = "../../../dataset_src/ds-1.tsv"
ds2_path = "../../../dataset_src/ds-2.tsv"
```

```python
timeline = [str(year) for year in range(2000, 2019)]
print(timeline)

spread_dict = dict()
```

The sequence of operations is:
- a graph for each year of timeline
- build a ds1 graph with fraction of co-citations weights calculation strategy
- application of the pagerank metric for each graph created
- spread of influence object for graph and metric result as seed for each year to apply the models

```python
def spread_timeline_k(k):
    for year in timeline:
        graph = gr.YearGraph(ds1_path, ds2_path, year)
        graph.read()
        graph.weights_fraction_cocitations()
        graph.build_ds1_graph()
        
        metrics = me.Metrics(graph)
        metrics.metrics_preprocessing()
        metrics.pagerank(k) 
    
        model = sp.SpreadOfInfluenceModels(graph, metrics.get_top_k_nodes())
        model.spread_preprocessing()
    
        spread_dict[year] = model
```

Function to print the results, iteration by iteration, of the Independent Cascade and Tipping models, applying the three threshold calculation strategies to each one.

```python
def to_print(influenced_nodes, name_model, k , threshold):
    comment = "Spread of influence result with {0} model, seed top {1} nodes and {2} threshold calculation strategy.\n".format(name_model, k, threshold)
    print(comment.upper())
    for key in influenced_nodes:
        print("{0} : {1}{2}len_iteration: {3}{4}".format(key, influenced_nodes[key], '\n', len(influenced_nodes[key]), '\n'))
    print("{0}".format('\n'))
```

Function for the calculation of the two models by applying to each one both the threshold calculation strategies and the printing of the respective results.

Regarding the nomenclature for threshold calculation strategies:

- 'deg threshold' is the threshold 1 of the report.
- 'half threshold' is the threshold 2 of the report.
- 'deg negative' is the threshold 3 of the report.

```python
def spread_view(year, k):
    spread_dict[str(year)].threshold_deg()
    spread_dict[str(year)].indipendent_cascade_model()
    cascade_deg = spread_dict[str(year)].get_influenced_nodes()
    to_print(cascade_deg, "cascade", k, "deg")
    
    spread_dict[str(year)].threshold_half()
    spread_dict[str(year)].indipendent_cascade_model()
    cascade_half = spread_dict[str(year)].get_influenced_nodes() 
    to_print(cascade_half, "cascade", k, "half")
    
    spread_dict[str(year)].threshold_deg_neg()
    spread_dict[str(year)].indipendent_cascade_model()
    cascade_deg = spread_dict[str(year)].get_influenced_nodes()
    to_print(cascade_deg, "cascade", k, "deg negative")
    
    print("-------------------------------------------------------------------------------------------\n\n\n")
        
    spread_dict[str(year)].threshold_deg()
    spread_dict[str(year)].tipping_model()
    tipping_deg = spread_dict[str(year)].get_influenced_nodes()   
    to_print(tipping_deg, "tipping", k, "deg")
     
    spread_dict[str(year)].threshold_half()
    spread_dict[str(year)].tipping_model()
    tipping_half = spread_dict[str(year)].get_influenced_nodes()
    to_print(tipping_half, "tipping", k, "half")
    
    spread_dict[str(year)].threshold_deg_neg()
    spread_dict[str(year)].tipping_model()
    tipping_deg = spread_dict[str(year)].get_influenced_nodes()   
    to_print(tipping_deg, "tipping", k, "deg negative")
```
