# Weights Strategies Analysis

This file shows the weights and its distribution to applying the two weight calculation strategies to modify the weight of the graph edges of the ds-1 data set along the timeline. The first strategy is the calculation of weights through the fraction of the co-citations made by the authors who have all worked together at least once between them; the second strategy is the number of co-citations for keywords pair (pair of nodes).

```python
import warnings
warnings.filterwarnings("ignore")

import sys

#Add path for class developed import
sys.path.append("..")

import graph as gr

ds1_path = '../../../dataset_src/ds-1.tsv'
ds2_path = '../../../dataset_src/ds-2.tsv'
```

Calculation of weights according to both strategies, edges-weights printing and plotting. 

```python
def year_weights_configs(year):
    graph = gr.YearGraph(ds1_path, ds2_path, year)
    graph.read()
    
    print("Weights config: fraction of co-citations keywords by cooperating authors.")
    graph.weights_fraction_cocitations()
    graph.plot_weights()
    print(graph.get_weights())
    
    print("{0}Weights config: number of co-citated keywords.".format('\n\n\n'))
    graph.weights_number_cocitations()
    graph.plot_weights()
    print(graph.get_weights())  
```
